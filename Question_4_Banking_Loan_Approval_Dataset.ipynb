Question 4: Banking Loan Approval Dataset

Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 2: Create the Dataset

```python
data = {
    "Customer_ID": ["C01","C02","C03","C04","C05","C06","C07","C08","C09","C10",
                    "C11","C12","C13","C14","C15","C16","C17","C18","C19","C20"],

    "Income": [45000,25000,60000,22000,50000,30000,70000,28000,55000,24000,
               65000,35000,48000,26000,75000,32000,52000,27000,68000,23000],

    "Credit_Score": [720,580,750,560,710,600,780,590,730,570,
                     760,620,700,585,790,610,725,575,770,555],

    "Loan_Amount": [200000,150000,300000,120000,250000,180000,350000,160000,270000,130000,
                    320000,200000,240000,140000,400000,190000,260000,150000,330000,125000],

    "Employment_Type": ["Salaried","Self-employed","Salaried","Unemployed","Salaried",
                        "Self-employed","Salaried","Self-employed","Salaried","Unemployed",
                        "Salaried","Self-employed","Salaried","Unemployed","Salaried",
                        "Self-employed","Salaried","Self-employed","Salaried","Unemployed"],

    "Loan_Status": ["Approved","Rejected","Approved","Rejected","Approved",
                    "Rejected","Approved","Rejected","Approved","Rejected",
                    "Approved","Rejected","Approved","Rejected","Approved",
                    "Rejected","Approved","Rejected","Approved","Rejected"]
}

df = pd.DataFrame(data)

print(df)
```

Step 3: Dataset Description

```python
print("Shape of Dataset:", df.shape)

print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nColumn Names")
print(df.columns)

print("\nData Types")
print(df.dtypes)
```

Step 4: Dataset Information

```python
print(df.info())
```

Step 5: Missing Values

```python
print("Missing Values")

print(df.isnull().sum())
```

Step 6: Duplicate Values

```python
print("Duplicate Rows:", df.duplicated().sum())
```

Step 7: Remove Duplicates

```python
df = df.drop_duplicates()

print("Shape after removing duplicates:", df.shape)
```

Step 8: Descriptive Statistics

```python
print(df.describe())
```

Step 9: Mean, Median and Standard Deviation

```python
print("Mean")

print(df.mean(numeric_only=True))

print("\nMedian")

print(df.median(numeric_only=True))

print("\nStandard Deviation")

print(df.std(numeric_only=True))
```

Step 10: Numerical Variable Analysis

```python
print(df[["Income","Credit_Score","Loan_Amount"]].describe())
```

Step 11: Categorical Variable Analysis

```python
print(df["Employment_Type"].value_counts())

print("\nLoan Status")

print(df["Loan_Status"].value_counts())
```

Step 12: Correlation Matrix

```python
print(df.corr(numeric_only=True))
```

Step 13: Histogram (Income Distribution)

```python
plt.figure(figsize=(6,4))

plt.hist(df["Income"], bins=5)

plt.title("Income Distribution")

plt.xlabel("Income")

plt.ylabel("Number of Customers")

plt.show()
```

Step 14: Bar Chart (Loan Status)

```python
df["Loan_Status"].value_counts().plot(kind="bar", color=["green","red"])

plt.title("Loan Approval Status")

plt.xlabel("Loan Status")

plt.ylabel("Count")

plt.show()
```

Step 15: Line Chart (Credit Score)

```python
plt.figure(figsize=(8,4))

plt.plot(df["Credit_Score"], marker='o')

plt.title("Credit Score of Customers")

plt.xlabel("Customer Index")

plt.ylabel("Credit Score")

plt.show()
```

Step 16: Scatter Plot (Income vs Loan Amount)

```python
plt.figure(figsize=(6,4))

plt.scatter(df["Income"], df["Loan_Amount"])

plt.title("Income vs Loan Amount")

plt.xlabel("Income")

plt.ylabel("Loan Amount")

plt.show()
```

Step 17: Box Plot (Credit Score)

```python
plt.figure(figsize=(5,4))

plt.boxplot(df["Credit_Score"])

plt.title("Credit Score Box Plot")

plt.show()
```

Approved Loans

```python
approved = df[df["Loan_Status"] == "Approved"]

print("Approved Loan Customers")

print(approved)
```

Rejected Loans

```python
rejected = df[df["Loan_Status"] == "Rejected"]

print("Rejected Loan Customers")

print(rejected)
```

Observations

The dataset contains 20 customer records and 6 columns.

There are no missing values.

There are no duplicate records.

Customers with higher income and higher credit scores are mostly approved for loans.

Most rejected applications belong to customers with lower income and lower credit scores.

Salaried customers have a higher number of approved loans than other employment types.

Conclusion

The Banking Loan Approval dataset indicates that income, credit score, and employment type play an important role in loan approval decisions. Customers with stable salaries, higher incomes, and better credit scores are more likely to receive loan approval. This analysis helps banks make informed lending decisions and identify applicants with higher approval chances.
