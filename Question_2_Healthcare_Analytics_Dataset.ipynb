Question 2: Healthcare Analytics Dataset

Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 2: Create the Dataset

```python
data = {
    "Patient_ID": ["P01","P02","P03","P04","P05","P06","P07","P08","P09","P10",
                   "P11","P12","P13","P14","P15","P16","P17","P18","P19","P20"],

    "Age": [45,32,55,28,60,40,52,35,48,30,
            62,38,50,27,58,42,46,34,53,29],

    "Sugar_Level": [145,98,180,90,200,120,165,105,155,95,
                    210,115,170,88,190,130,150,100,175,92],

    "BP": [130,118,145,110,150,125,140,120,135,115,
           155,122,142,108,148,128,132,116,144,112],

    "BMI": [28,23,31,22,34,26,30,24,29,23,
            35,25,30,21,33,27,28,24,31,22],

    "Disease_Status": ["Yes","No","Yes","No","Yes","No","Yes","No","Yes","No",
                       "Yes","No","Yes","No","Yes","No","Yes","No","Yes","No"]
}

df = pd.DataFrame(data)

print(df)
```

Step 3: Dataset Description

```python
print("Shape of Dataset:", df.shape)

print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nColumn Names")
print(df.columns)

print("\nData Types")
print(df.dtypes)
```

Step 4: Dataset Information

```python
print(df.info())
```

Step 5: Missing Values

```python
print("Missing Values")

print(df.isnull().sum())
```

Step 6: Duplicate Values

```python
print("Duplicate Rows:", df.duplicated().sum())
```

Step 7: Remove Duplicates

```python
df = df.drop_duplicates()

print("Shape after removing duplicates:", df.shape)
```

Step 8: Descriptive Statistics

```python
print(df.describe())
```

Step 9: Mean, Median, Standard Deviation

```python
print("Mean")

print(df.mean(numeric_only=True))

print("\nMedian")

print(df.median(numeric_only=True))

print("\nStandard Deviation")

print(df.std(numeric_only=True))
```

Step 10: Numerical Variable Analysis

```python
print(df[["Age","Sugar_Level","BP","BMI"]].describe())
```

Step 11: Categorical Variable Analysis

```python
print(df["Disease_Status"].value_counts())
```

Step 12: Correlation Matrix

```python
print(df.corr(numeric_only=True))
```

Step 13: Histogram

```python
plt.figure(figsize=(6,4))

plt.hist(df["Sugar_Level"], bins=5)

plt.title("Distribution of Sugar Level")

plt.xlabel("Sugar Level")

plt.ylabel("Number of Patients")

plt.show()
```

Step 14: Bar Chart

```python
df["Disease_Status"].value_counts().plot(kind="bar", color=["red","green"])

plt.title("Disease Status")

plt.xlabel("Disease")

plt.ylabel("Count")

plt.show()
```

Step 15: Line Chart

```python
plt.figure(figsize=(8,4))

plt.plot(df["Age"], marker='o')

plt.title("Age of Patients")

plt.xlabel("Patient Index")

plt.ylabel("Age")

plt.show()
```

Step 16: Scatter Plot

```python
plt.figure(figsize=(6,4))

plt.scatter(df["Sugar_Level"], df["BMI"])

plt.title("Sugar Level vs BMI")

plt.xlabel("Sugar Level")

plt.ylabel("BMI")

plt.show()
```

Observations

The dataset contains 20 patient records and 6 columns.

There are no missing values in the dataset.

There are no duplicate records.

Patients with higher sugar levels generally have higher BMI and blood pressure.

Patients with high sugar levels are more likely to have Disease_Status = Yes.

Age, blood pressure, and BMI show a positive relationship with disease status.

Conclusion

The Healthcare Analytics dataset shows that patients with higher sugar levels, higher blood pressure, and higher BMI are more likely to have the disease. This EDA helps hospitals identify high-risk patients and supports early diagnosis and treatment decisions.
