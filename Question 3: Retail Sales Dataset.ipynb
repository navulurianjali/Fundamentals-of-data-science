Question 3: Retail Sales Dataset

Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 2: Create the Dataset

```python
data = {
    "Product_ID": ["PR01","PR02","PR03","PR04","PR05","PR06","PR07","PR08","PR09","PR10",
                   "PR11","PR12","PR13","PR14","PR15","PR16","PR17","PR18","PR19","PR20"],

    "Category": ["Grocery","Snacks","Dairy","Fruits","Vegetables",
                 "Grocery","Snacks","Dairy","Fruits","Vegetables",
                 "Grocery","Snacks","Dairy","Fruits","Vegetables",
                 "Grocery","Snacks","Dairy","Fruits","Vegetables"],

    "Price": [50,30,45,60,40,80,25,55,70,35,
              90,20,65,75,45,100,35,50,80,30],

    "Quantity_Sold": [20,35,18,25,30,15,40,16,22,28,
                      12,45,14,20,26,10,32,19,18,35],

    "Discount": [5,3,2,5,4,6,3,2,5,4,
                 6,2,3,5,4,7,3,2,6,3],

    "Revenue": [1000,1050,810,1500,1200,1200,1000,880,1540,980,
                1080,900,910,1500,1170,1000,1120,950,1440,1050]
}

df = pd.DataFrame(data)

print(df)
```

Step 3: Dataset Description

```python
print("Shape of Dataset:", df.shape)

print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nColumn Names")
print(df.columns)

print("\nData Types")
print(df.dtypes)
```

Step 4: Dataset Information

```python
print(df.info())
```

Step 5: Missing Values

```python
print("Missing Values")
print(df.isnull().sum())
```

Step 6: Duplicate Values

```python
print("Duplicate Rows:", df.duplicated().sum())
```

Step 7: Remove Duplicates

```python
df = df.drop_duplicates()

print("Shape after removing duplicates:", df.shape)
```

Step 8: Descriptive Statistics

```python
print(df.describe())
```

Step 9: Mean, Median, Standard Deviation

```python
print("Mean")
print(df.mean(numeric_only=True))

print("\nMedian")
print(df.median(numeric_only=True))

print("\nStandard Deviation")
print(df.std(numeric_only=True))
```

Step 10: Numerical Variable Analysis

```python
print(df[["Price","Quantity_Sold","Discount","Revenue"]].describe())
```

Step 11: Categorical Variable Analysis

```python
print(df["Category"].value_counts())
```

Step 12: Correlation Matrix

```python
print(df.corr(numeric_only=True))
```

Step 13: Histogram (Revenue Distribution)

```python
plt.figure(figsize=(6,4))

plt.hist(df["Revenue"], bins=5)

plt.title("Revenue Distribution")

plt.xlabel("Revenue")

plt.ylabel("Number of Products")

plt.show()
```

Step 14: Bar Chart (Products in Each Category)

```python
df["Category"].value_counts().plot(kind="bar", color=["blue","green","orange","red","purple"])

plt.title("Number of Products in Each Category")

plt.xlabel("Category")

plt.ylabel("Count")

plt.show()
```

Step 15: Line Chart (Revenue Trend)

```python
plt.figure(figsize=(8,4))

plt.plot(df["Revenue"], marker='o')

plt.title("Revenue of Products")

plt.xlabel("Product Index")

plt.ylabel("Revenue")

plt.show()
```

Step 16: Scatter Plot (Price vs Revenue)

```python
plt.figure(figsize=(6,4))

plt.scatter(df["Price"], df["Revenue"])

plt.title("Price vs Revenue")

plt.xlabel("Price")

plt.ylabel("Revenue")

plt.show()
```

High-Selling Products

```python
print("Top 5 High-Selling Products")

print(df.sort_values(by="Quantity_Sold", ascending=False).head())
```

Highest Revenue Products

```python
print("Top 5 Revenue Generating Products")

print(df.sort_values(by="Revenue", ascending=False).head())
```

Observations

The dataset contains 20 products and 6 columns.

There are no missing values.

There are no duplicate records.

Snacks have the highest quantity sold.

Fruits generate the highest revenue among the categories.

Revenue generally increases with product price, though quantity sold also affects total revenue.

Conclusion

The Retail Sales dataset shows that different product categories contribute differently to sales and revenue. Fruits generate high revenue, while snacks have the highest sales volume. This analysis helps supermarket management identify high-selling products, optimize inventory, and make better pricing and promotional decisions.
